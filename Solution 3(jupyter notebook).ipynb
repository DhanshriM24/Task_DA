# Analysis Of Data

#### 1.Reward the top performer (student) of each school based on cumulative marks scored in last three years for all the subjects

```python
# Calculate cumulative marks
subjects = ['Hindi', 'English', 'Mathematics', 'Physics', 'Chemistry', 'Biology', 'History', 'Geography', 'Civics', 'Computer Science', 'Physical Education']
data['Total_Marks'] = data[subjects].sum(axis=1)

# Find the top performer in each school
top_performers = data.groupby('School')['Total_Marks'].idxmax()
top_performers_df = data.loc[top_performers]

top_performers_df[['School', 'Student Roll', 'Student Name', 'Total_Marks']]
```

```python
# Bar chart for top performers
plt.figure(figsize=(6, 6))
plt.bar(top_performers_df['School'], top_performers_df['Total_Marks'])
plt.xlabel('School')
plt.ylabel('Total Marks')
plt.title('Top Performer in Each School')
plt.show()
```

#### 2.Rank each student within their own school based on their total marks scored in the year 2020 and compare the marks of Rank 10 for each school by arranging them in descending order

```python
data_2020 = data[data['Year'] == 2020]
data_2020['Total_Marks'] = data_2020[subjects].sum(axis=1)

# Rank students within each school based on total marks
data_2020['Rank'] = data_2020.groupby('School')['Total_Marks'].rank(method='min', ascending=False)

# Extract the 10th ranked student
rank_10_2020 = data_2020[data_2020['Rank'] == 10]

# Compare these marks across schools
rank_10_comparison = rank_10_2020[['School', 'Total_Marks','Student Name']].sort_values(by='Total_Marks', ascending=False)
rank_10_comparison
```

```python
plt.figure(figsize=(6, 6))
plt.bar(rank_10_comparison['School'], rank_10_comparison['Total_Marks'])
plt.xlabel('School')
plt.ylabel('Total Marks')
plt.title('Total Marks of 10th Ranked Students Across Schools (2020)')
plt.show()
```

#### 3.Find out students with the highest improvement for each subject from 2019-21 combining all the schools together

```python
top_improvers = {}

for subject in subjects:
    subject_pivot = data.pivot_table(index=['Student Name', 'Student Roll', 'School'], columns='Year', values=subject, aggfunc='mean')
    subject_pivot['Improvement'] = subject_pivot[2021] - subject_pivot[2019]
    
    top_improver = subject_pivot['Improvement'].idxmax()
    top_improvers[subject] = top_improver

print("Students with the highest improvement for each subject:")
for subject, (student_name, student_roll, school) in top_improvers.items():
    print(f"{subject}:")
    print(f"  Student Name: {student_name}")
    print(f"  Roll No: {student_roll}")
    print(f"  School: {school}")
    print()
```

#### 4.Identify best school for Arts, Science and Commerce streams based on marks scored by students in respective subjects for those streams in last three years

```python
streams = {
    'Arts': ['Hindi', 'English', 'History', 'Geography', 'Civics'],
    'Science': ['Mathematics', 'Physics', 'Chemistry', 'Biology', 'Computer Science'],
    'Commerce': ['Hindi', 'English', 'Mathematics', 'Computer Science', 'Physical Education']
}

# Calculate average marks for each stream
best_schools = {}
for stream, subjects in streams.items():
    avg_marks = data.groupby('School')[subjects].mean().mean(axis=1)
    best_school = avg_marks.idxmax()
    best_schools[stream] = best_school

best_schools
```

#### 5. If the marks obtained for each subject can be categorised under 5 sections like below:
####          <=20 - Very Poor
####    >20 - <=40 - Poor
####    >40 - <=60 - Average
####    >60 - <=80 - Good
####    >80 - <=100 - Very Good
#### Then calculate for each school how many students were in each category based on the avg. marks obtained each year

```python
def categorize(marks):
    if marks <= 20:
        return 'Very Poor'
    elif marks <= 40:
        return 'Poor'
    elif marks <= 60:
        return 'Average'
    elif marks <= 80:
        return 'Good'
    else:
        return 'Very Good'

# Calculate average marks for each stream
data['Avg_Marks'] = data[subjects].mean(axis=1)
data['Category'] = data['Avg_Marks'].apply(categorize)

category_counts = data.groupby(['School', 'Category']).size().unstack().fillna(0)
category_counts
```

```python
category_counts.plot(kind='bar', stacked=True, figsize=(5, 5))
plt.xlabel('School')
plt.ylabel('Number of Students')
plt.title('Student Distribution by Performance Category Across Schools')
plt.show()
```

#### 6. Which is the best school for each year 2019, 2020 and 2021 based on highest no. of students in Good and Very Good category

```python
best_schools = {}

for year in data['Year'].unique():
    data_year = data[data['Year'] == year]
    
    category_counts = data_year.groupby(['School', 'Category']).size().unstack().fillna(0)
    good_very_good = category_counts[['Good', 'Very Good']].sum(axis=1)
    
    best_school_by_category = good_very_good.idxmax()
    best_schools[year] = (best_school_by_category, good_very_good[best_school_by_category])

for year, (school, count) in best_schools.items():
    print(f"Year {year}: Best school is {school} with {count} students in 'Good' and 'Very Good' categories.")
```

#### 7. Which is the fastest-growing School in Bangalore (Overall and Streamwise)?

```python
fastest_growing_school_streamwise = {}

for stream, subjects in streams.items():
    data_stream = data[['School', 'Year'] + subjects]
    avg_marks_stream = data_stream.groupby(['School', 'Year'])[subjects].mean().reset_index()
    avg_marks_pivot = avg_marks_stream.pivot_table(index='School', columns='Year', values=subjects, aggfunc='mean')
    avg_marks_pivot.columns = [f'{col}_{year}' for year in [2019, 2020, 2021] for col in subjects]
    avg_marks_pivot['Growth'] = avg_marks_pivot[[f'{subj}_2021' for subj in subjects]].mean(axis=1) - avg_marks_pivot[[f'{subj}_2019' for subj in subjects]].mean(axis=1)
    fastest_growing_school_streamwise[stream] = avg_marks_pivot['Growth'].idxmax()

# Display results
print("Fastest-growing school by stream:")
for stream, school in fastest_growing_school_streamwise.items():
    print(f"{stream}: {school}")
```
